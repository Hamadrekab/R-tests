1. Data Extraction

 USe the Open Meteo API (https://open-meteo.com/) to extract weather data 


Use the historical forecast API  (https://open-meteo.com/en/docs/historical-forecast-api) to extract historical weather data

 Data parameters:

* Extract data for the last 90 days
* Extract data for Riyadh, London, California, Tokyo, and Sydney
* Extract the following hourly variables: Temperature, Relative Humidity, Precipitation Probability, Precipitation (rain + showers + snow), and Wind Gusts (10m)
* Extract the following daily variables: Weather code, Maximum temperature, Minimum temperature, UV Index


```python
import requests
import datetime
import pandas as pd
import json

# Base URL for the Open Meteo Historical Forecast API
base_url = "https://archive-api.open-meteo.com/v1/archive"

# Define the locations with their coordinates
locations = {
    "Riyadh": {"latitude": 24.7136, "longitude": 46.6753},
    "London": {"latitude": 51.5074, "longitude": -0.1278},
    "California": {"latitude": 36.7783, "longitude": -119.4179},
    "Tokyo": {"latitude": 35.6895, "longitude": 139.6917},
    "Sydney": {"latitude": -33.8688, "longitude": 151.2093}
}

# Define the date range (last 90 days)
end_date = datetime.date.today()
start_date = end_date - datetime.timedelta(days=90)

# Define the hourly and daily variables to extract
hourly_variables = [
    "temperature_2m", "relative_humidity_2m", "precipitation_probability",
    "precipitation", "wind_gusts_10m"
]
daily_variables = [
    "weathercode", "temperature_2m_max", "temperature_2m_min", "uv_index_max"
]
```

```python
def fetch_weather_data(location, latitude, longitude):
    # Construct the request URL
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": start_date.isoformat(),
        "end_date": end_date.isoformat(),
        "hourly": ",".join(hourly_variables),
        "daily": ",".join(daily_variables),
        "timezone": "auto"
    }
    
    # Make the request
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to retrieve data for {location}. Status code: {response.status_code}")
        return None

# Fetch data for all locations
weather_data = {}
for location, coords in locations.items():
    data = fetch_weather_data(location, coords["latitude"], coords["longitude"])
    if data:
        weather_data[location] = data

# Check a sample of the fetched data
print(weather_data['Riyadh']['hourly'].keys())  # Just to check the structure
```

2. Data Cleaning and Transformation

Data cleaning:
- Identify and handle missing or null values
- Detect and remove duplicate records
- Check for inconsistencies in data types and correct them
-Ensure all timestamps are in a consistent time zone and format

Transformation:
- Transform the data, so that it is in tabular format


```python
def clean_weather_data(raw_data):
    # Convert the raw JSON data to a DataFrame
    hourly_data_list = []
    daily_data_list = []
    
    for location, data in raw_data.items():
        # Extract hourly data
        hourly_df = pd.DataFrame(data['hourly'])
        hourly_df['location'] = location
        hourly_data_list.append(hourly_df)
        
        # Extract daily data
        daily_df = pd.DataFrame(data['daily'])
        daily_df['location'] = location
        daily_data_list.append(daily_df)
    
    # Concatenate all location data into a single DataFrame
    hourly_data = pd.concat(hourly_data_list, ignore_index=True)
    daily_data = pd.concat(daily_data_list, ignore_index=True)
    
    # Convert timestamps to datetime
    hourly_data['time'] = pd.to_datetime(hourly_data['time'])
    daily_data['time'] = pd.to_datetime(daily_data['time'])
    
    # Handle missing values (fill forward)
    hourly_data.ffill(inplace=True)
    daily_data.ffill(inplace=True)
    
    # Remove duplicates
    hourly_data.drop_duplicates(inplace=True)
    daily_data.drop_duplicates(inplace=True)
    
    return hourly_data, daily_data

# Clean the weather data
hourly_data, daily_data = clean_weather_data(weather_data)

# Display a sample of the cleaned data
print(hourly_data.head())
print(daily_data.head())
```

```python
def transform_data(hourly_data, daily_data):
    # Ensure consistent column names
    hourly_data.columns = hourly_data.columns.str.lower()
    daily_data.columns = daily_data.columns.str.lower()
    
    return hourly_data, daily_data

# Transform the data
hourly_data, daily_data = transform_data(hourly_data, daily_data)

# Display transformed data
hourly_data.head()
```

3. Data Storage
- Save the cleaned dataset in Parquet format
- Partition the Parquet file by date to optimise for efficient querying and storage


```python
import pyarrow  # Ensure pyarrow is installed

def save_data_as_parquet(hourly_data, daily_data):
    # Add a date column for partitioning
    hourly_data['date'] = hourly_data['time'].dt.date
    daily_data['date'] = daily_data['time'].dt.date

    try:
        # Save hourly data partitioned by location and date
        hourly_data.to_parquet('hourly_weather_data.parquet', partition_cols=['location', 'date'], engine='pyarrow', index=False)
        print("Hourly data saved successfully as 'hourly_weather_data.parquet'")
        
        # Save daily data partitioned by location and date
        daily_data.to_parquet('daily_weather_data.parquet', partition_cols=['location', 'date'], engine='pyarrow', index=False)
        print("Daily data saved successfully as 'daily_weather_data.parquet'")
    except Exception as e:
        print(f"An error occurred while saving data: {e}")

# Save the data in Parquet format
save_data_as_parquet(hourly_data, daily_data)
```

```python
import pandas as pd

# Load the hourly and daily data from Parquet files
hourly_data_loaded = pd.read_parquet('hourly_weather_data.parquet')
daily_data_loaded = pd.read_parquet('daily_weather_data.parquet')

# Display the first few rows to verify the data is loaded correctly
print(hourly_data_loaded.head())
print(daily_data_loaded.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
sns.set_theme(style="whitegrid")

# Plot the temperature over time for each location
plt.figure(figsize=(14, 7))
sns.lineplot(data=hourly_data_loaded, x='time', y='temperature_2m', hue='location', palette='viridis')
plt.title('Temperature Trends Over Time by Location', fontsize=16)
plt.xlabel('Time', fontsize=14)
plt.ylabel('Temperature (C)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Location', loc='upper right')
plt.show()
```

```python
# Plot daily max and min temperatures for each location
plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_data_loaded, x='time', y='temperature_2m_max', hue='location', palette='coolwarm', marker='o', label='Max Temperature')
sns.lineplot(data=daily_data_loaded, x='time', y='temperature_2m_min', hue='location', palette='coolwarm', marker='o', linestyle='--', label='Min Temperature')
plt.title('Daily Maximum and Minimum Temperatures by Location', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Temperature (C)', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Temperature', loc='upper left')
plt.show()
```

```python
# Pivot the data to create a heatmap-friendly format for temperature in Riyadh
riyadh_temp = hourly_data_loaded[hourly_data_loaded['location'] == 'Riyadh'].pivot_table(
    index=hourly_data_loaded['time'].dt.date, 
    columns=hourly_data_loaded['time'].dt.hour, 
    values='temperature_2m'
)

# Plot the heatmap for temperature in Riyadh
plt.figure(figsize=(16, 8))
sns.heatmap(riyadh_temp, cmap='coolwarm', cbar_kws={'label': 'Temperature (C)'}, linewidths=.5)
plt.title('Temperature Heatmap - Riyadh', fontsize=16)
plt.xlabel('Hour of the Day', fontsize=14)
plt.ylabel('Date', fontsize=14)
plt.show()
```
